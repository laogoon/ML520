INDUS：城镇非零售业商用土地比例
CHAS：是否邻近查尔斯河，1是邻近，0是不邻近
NOX：一氧化氮浓度（千万分之一）

MEDV：自住房中位数价格，单位是千元

```python
from sklearn.datasets import load_boston
import pandas as pd
boston = load_boston()
df = pd.DataFrame(boston["data"], columns=boston["feature_names"])
df["ans"] = boston["target"]
# df.to_csv("boston.csv", encoding="utf-8", index=False)
df
```

```python
# 把資料{隨機}分成兩組 - 一組(90%)訓練，一組考試(10%)
from sklearn.model_selection import train_test_split
# [90%題目, 10%題目, 90%答案, 10%答案]
x_train, x_test, y_train, y_test = train_test_split(boston["data"],
                                                    boston["target"],
                                                    test_size=0.1)
```

```python
# 準備模型
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(max_depth=5)
```

```python
# 訓練(fit)模型: 考古題
reg.fit(x_train, y_train)
```

```python
# 秀出公式長啥樣子
#!apt-get -qq instll -y graphviz && pip instll -q graphviz
import graphviz
from sklearn.tree import export_graphviz
g = export_graphviz(reg,
                    feature_names=boston["feature_names"],
                    filled=True)
graphviz.Source(g)
```

```python
# 驗證模型: 考試
from sklearn.metrics import r2_score
pre = reg.predict(x_test)
print("預測:", list(pre))
print("正確:", list(y_test))
print("R2 Score:", r2_score(y_test, pre))
```
