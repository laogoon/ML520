```python
import pandas as pd
train_df = pd.read_csv("train.csv", encoding="utf-8")
train_df
```

```python
test_df = pd.read_csv("test.csv", encoding="utf-8")
test_df
```

```python
# Step1.補缺失值
train_df.isna().sum()
```

```python
med = train_df.median()
train_df = train_df.fillna(med)
# train_df["行名稱"] = train_df["行名稱"].fillna(0.006)
train_df.isna().sum()
```

```python
test_df.isna().sum()
```

```python
# 注意：不要對測試資料重新算median
test_df = test_df.fillna(med)
test_df.isna().sum()
```

```python
freq = train_df["Embarked"].value_counts().idxmax()
train_df["Embarked"] = train_df["Embarked"].fillna(freq)
train_df.isna().sum()
```

```python
test_df["Embarked"] = test_df["Embarked"].fillna(freq)
test_df.isna().sum()
```

```python
# Step2. One-Hot Encoding
dummy = pd.get_dummies(train_df["Embarked"])
train_df = pd.concat([train_df, dummy], axis=1)
train_df
```

```python
dummy = pd.get_dummies(test_df["Embarked"])
test_df = pd.concat([test_df, dummy], axis=1)
test_df
```

```python
dummy = pd.get_dummies(train_df["Sex"])
train_df = pd.concat([train_df, dummy], axis=1)
train_df
```

```python
dummy = pd.get_dummies(test_df["Sex"])
test_df = pd.concat([test_df, dummy], axis=1)
test_df
```

```python
# pandas: apply
def middle(s):
    mid = s.split(",")[-1].split(".")[0].replace(" ", "")
    reserved = ["Mr", "Mrs", "Miss"]
    if mid in reserved:
        return mid
    else:
        return "X"
dummy = pd.get_dummies(train_df["Name"].apply(middle))
train_df = pd.concat([train_df, dummy], axis=1)
train_df
```

```python
dummy = pd.get_dummies(test_df["Name"].apply(middle))
test_df = pd.concat([test_df, dummy], axis=1)
test_df
```

```python
train = train_df.drop(['PassengerId', 'Name', 'Sex', 
                       'Ticket', 'Cabin', 'Embarked',
                       'X'], axis=1)
x_train = train.drop(['Survived'], axis=1)
y_train = train["Survived"]
x_train
```
